Simple algorithm for dependence checking

```python
import numpy as np
from IPython.display import Latex, display

# Function to check if two vectors are dependent
def are_vectors_dependent(vector1, vector2):
    # Convert the vectors to NumPy arrays
    v1 = np.array(vector1)
    v2 = np.array(vector2)

    # Check if one vector is a scalar multiple of the other
    for i in range(len(v1)):
        if v1[i] != 0 and v2[i] != 0:
            if v1[i] / v2[i] == v1[0] / v2[0]:
                continue
            else:
                return False
        elif v1[i] == 0 and v2[i] != 0:
            return False
        elif v2[i] == 0 and v1[i] != 0:
            return False

    return True

# Example vectors
vector1 = [2, 4]
vector2 = [6, 8]
vector3 = [1, 2]

# Check dependence
result1 = are_vectors_dependent(vector1, vector2)
result2 = are_vectors_dependent(vector1, vector3)

# Print LaTeX code for vectors
display(Latex(f"$\\large Vector_1:$ $\\large \\mathbf{{v}}_1 = \\begin{{bmatrix}}{vector1[0]} \\\\ {vector1[1]}\\end{{bmatrix}}$"))
display(Latex(f"$\\large Vector_2:$ $\\large \\mathbf{{v}}_2 = \\begin{{bmatrix}}{vector2[0]} \\\\ {vector2[1]}\\end{{bmatrix}}$"))
display(Latex(f"The are {'not' if not result1 else ''} vectors dependent"))

display(Latex(f"$\\large Vector_1:$ $\\large\\mathbf{{v}}_1 = \\begin{{bmatrix}}{vector1[0]} \\\\ {vector1[1]}\\end{{bmatrix}}$"))
display(Latex(f"$\\large Vector_3:$ $\\large\\mathbf{{v}}_3 = \\begin{{bmatrix}}{vector3[0]} \\\\ {vector3[1]}\\end{{bmatrix}}$"))
display(Latex(f"The are {'not' if not result2 else ''} vectors dependent"))
```

```python
import numpy as np

# Define a square matrix with integer values
A = np.array([[1, 2, 1],
              [4, 4, 5],
              [6, 7, 7]])

# Check if the matrix is invertible (has a non-zero determinant)
det_A = np.linalg.det(A)
if det_A != 0:
    # Calculate the inverse matrix
    A_inv = np.linalg.inv(A)

    # Verify that A * A_inv equals the identity matrix
    identity_matrix = np.dot(A, A_inv)

    print("Original Matrix A:")
    print(A)
    print("\nInverse Matrix A_inv:")
    print(A_inv)
    print("\nIdentity Matrix:")
    print(np.round(identity_matrix))  # Round to handle floating-point precision

    # Verify that A * A_inv is approximately equal to the identity matrix
    print("\nVerification (A * A_inv is approximately equal to the Identity Matrix):")
    print(np.allclose(identity_matrix, np.eye(A.shape[0], dtype=int)))
else:
    print("Matrix A is not invertible since its determinant is zero.")
```

```python
# Check if the matrix is upper/lower triangular
import numpy as np

def is_upper_triangular(matrix):
    return np.all(np.triu(matrix) == matrix)

def is_lower_triangular(matrix):
    return np.all(np.tril(matrix) == matrix)

# Example matrix
matrix = np.array([[1, 2, 3, 7],
                   [0, 4, 5, 7],
                   [0, 0, 0, 1]])

# Check if the matrix is upper triangular
if is_upper_triangular(matrix):
    print("The matrix is upper triangular.")
elif is_lower_triangular(matrix):
    print("The matrix is lower triangular.")
else:
    print("The matrix is neither upper nor lower triangular.")
```

Now consider the system of two linear equations

$$
\begin{cases}
x - 3y = -3 \\
2x + y = 8
\end{cases}
$$

Each equation individually defines a line in the plane, pictured below. 

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import Matrix, latex
from IPython.display import display, Latex

plt.title('Two equations system', fontsize=16)
plt.ylabel('y', fontsize=14)
plt.xlabel('x', fontsize=14)
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.plot([0, 0], [-10, 10], 'k')
plt.plot([10, -10], [0, 0], 'k')

# Plot the first equation graph
x1 = np.linspace(-10, 10, 100)
y1 = x1/3 + 1

plt.plot(x1, y1)

# Plot the second equation graph
x2 = np.linspace(-10, 10, 100)
y2 = 8 - 2*x2

plt.plot(x2, y2)

# Calculate and plot the solution point
A = np.matrix([[1, -3], [2, 1]])
b = np.array([-3, 8])

x = np.linalg.solve(A, b)

plt.plot(x[0], x[1], marker='o')
plt.text(x[0], x[1]+1, '({}, {})'.format(x[0], x[1]))
plt.grid()


import numpy as np
from sympy import *
from IPython.display import Latex

# Define a square matrix with integer values
A = Matrix(A)
b = Matrix(b)

sol, params = A.gauss_jordan_solve(b)

display(Latex("$\Large " + latex(A) + " \cdot " + latex(Matrix(["x_{}".format(i) for i in range(1, A.shape[1]+1)])) + " = " + latex(b) +"$"))
display(Latex("$\Large " + latex(Matrix(["x_{}".format(i) for i in range(1, A.shape[1]+1)])) + " = " + latex(sol) + "$"))
```

A solution to the system of both equations is a pair of numbers $( x , y )$ that makes both equations true at once. In other words, it as a point that lies on both lines simultaneously. We can see in the picture above that there is only one point where the lines intersect: therefore, this system has exactly one solution.

Usually, two lines in the plane will intersect in one point, but of course this is not always the case. Consider now the system of equations

$$
\begin{cases}
x - 3y = -3 \\
x - 3y = 8
\end{cases}
$$

These define parallel lines in the plane.

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import Matrix, latex
from IPython.display import display, Latex

plt.title('Two equations system', fontsize=16)
plt.ylabel('y', fontsize=14)
plt.xlabel('x', fontsize=14)
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.plot([0, 0], [-10, 10], 'k')
plt.plot([10, -10], [0, 0], 'k')

# Plot the first equation graph
x1 = np.linspace(-10, 10, 100)
y1 = x1/3 + 1

plt.plot(x1, y1)

# Plot the second equation graph
x2 = np.linspace(-10, 10, 100)
y2 = (x2 - 8)/3

plt.plot(x2, y2)
plt.grid()
plt.show()
```

The fact that that the lines do not intersect means that the system of equations has no solution. Of course, this is easy to see algebraically: if $x  3y =  3$,
then it is cannot also be the case that $x  3 y = 3$.

\end{cases}

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Latex, display

# Consider the vector v in the vector space V
v = sp.Matrix([3, 2])

# Consider two basis of the vector space V
# B basis
b1 = sp.Matrix([-1, -1])
b2 = sp.Matrix([2, 1])

# B' basis
b1_prime = sp.Matrix([1, 2])
b2_prime = sp.Matrix([1, -1])

# The coordinates of v in the B basis are
v_b_coordinates = b1 * v[0] + b2 * v[1]

# The coordinates of v in the B' basis are
v_b_prime_coordinates = b1_prime * v[0] + b2_prime * v[1]

# Display the results

print("The coordinates of v in the canonical basis are:")
display(Latex("$[\mathbf v ]_E =" + sp.latex(v) + "$"))
display(Latex("$[\mathbf v ]_B =" + sp.latex(v_b_coordinates) + "$"))
display(Latex("$[\mathbf v ]_{B'} =" + sp.latex(v_b_prime_coordinates) + "$"))

# The matrix from E to B is
matrix_E_to_B = sp.Matrix([
    [-1, 2],
    [-1, 1]
])

# The matrix from E to B' is
matrix_E_to_B_prime = sp.Matrix([
    [1, 1],
    [2, -1]
])

# Print the matrixes
print("The matrix from E to B is:")
display(Latex("$\mathbf E_B =" + sp.latex(matrix_E_to_B) + "$"))
print("The matrix from E to B' is:")
display(Latex("$\mathbf E_{B'} =" + sp.latex(matrix_E_to_B_prime) + "$"))

# Infact if we multiply the inversematrices by the vector's respective coordinates
# we get the initial vector
print("The vector in the initial basis is:")
display(Latex("$\mathbf v = \mathbf E_B^{-1}" + sp.latex(v_b_coordinates) + " = " + sp.latex(matrix_E_to_B.inv() * v_b_coordinates) + "$"))
display(Latex("$\mathbf v = \mathbf E_{B'}^{-1}" + sp.latex(v_b_prime_coordinates) + " = " + sp.latex(matrix_E_to_B_prime.inv() * v_b_prime_coordinates) + "$"))

# If we want to map the vector in the B basis to the B' basis, we have to multiply the
# inverse of the matrix from E to B by the matrix from E to B' and then multiply by the vector's respective coordinates

# The matrix from B to B' is
matrix_B_to_B_prime = matrix_E_to_B_prime * matrix_E_to_B.inv()

# Print the matrix
print("The matrix from B to B' is:")
display(Latex("$\mathbf B_{B'} =" + sp.latex(matrix_B_to_B_prime) + "$"))

# Infact if we multiply the inversematrices by the vector's respective coordinates
display(Latex("$[\mathbf v ]_{B'} = \mathbf B_{B'}" + sp.latex(v_b_coordinates) + " = " + sp.latex(matrix_B_to_B_prime * v_b_coordinates) + "$"))

# The matrix from B' to B is
matrix_B_prime_to_B = matrix_E_to_B * matrix_E_to_B_prime.inv()

# Print the matrix
print("The matrix from B' to B is:")
display(Latex("$\mathbf {B'}_B =" + sp.latex(matrix_B_prime_to_B) + " = \mathbf B_{B'}^{-1} = " + sp.latex(matrix_B_to_B_prime.inv()) + "$"))

# Infact if we multiply the inversematrices by the vector's respective coordinates
display(Latex("$[\mathbf v ]_B = \mathbf {B'}_B" + sp.latex(v_b_prime_coordinates) + " = " + sp.latex(matrix_B_prime_to_B * v_b_prime_coordinates) + "$"))
```
